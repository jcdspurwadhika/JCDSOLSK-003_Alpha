### Error Type Consideration & Evaluation Metric

Dalam konteks prediksi churn, kelas positif didefinisikan sebagai pelanggan yang churn (Churn = 1), sedangkan kelas negatif adalah pelanggan yang tidak churn (Churn = 0).

**Type I Error (False Positive)**  
Model memprediksi pelanggan akan churn, padahal sebenarnya tidak.  
Konsekuensi: perusahaan memberikan insentif atau program retensi kepada pelanggan yang sebenarnya tidak berisiko churn. Dampaknya berupa biaya tambahan untuk program retensi.

**Type II Error (False Negative)**  
Model memprediksi pelanggan tidak akan churn, padahal sebenarnya churn.  
Konsekuensi: perusahaan gagal melakukan intervensi kepada pelanggan berisiko tinggi, sehingga berpotensi kehilangan pendapatan dan harus mengeluarkan biaya akuisisi pelanggan baru.

Dalam permasalahan churn, baik False Positive maupun False Negative sama-sama menimbulkan biaya bagi perusahaan. Oleh karena itu, evaluasi model tidak hanya berfokus pada satu jenis error saja.  

Metrik utama yang digunakan adalah **F1-score**, karena menggabungkan precision dan recall sehingga mampu menangkap trade-off antara False Positive dan False Negative secara bersamaan. Selain itu, confusion matrix, precision, dan recall tetap disajikan untuk memberikan gambaran yang lebih komprehensif terhadap performa model.

## 11. Data Preprocessing & Cleansing

Pada tahap ini dilakukan penanganan missing values serta pemeriksaan akhir terhadap kualitas data sebelum masuk ke tahap feature engineering dan modeling. Strategi preprocessing ditentukan berdasarkan hasil EDA dan karakteristik masing-masing variabel.


```python
# Make a copy of dataset
df_clean = df.copy()

# Drop CustomerID
df_clean.drop(columns=["CustomerID"], inplace=True)

# Separate numerical and categorical columns
num_cols = df_clean.select_dtypes(include=["int64","float64"]).columns
cat_cols = df_clean.select_dtypes(include=["object"]).columns

# Fill missing values
for col in num_cols:
    df_clean[col].fillna(df_clean[col].median(), inplace=True)

for col in cat_cols:
    df_clean[col].fillna(df_clean[col].mode()[0], inplace=True)

# Check remaining missing values
df_clean.isnull().sum()
```

### Preprocessing Result

Missing values pada seluruh variabel telah ditangani menggunakan median untuk variabel numerik dan mode untuk variabel kategorikal. Pendekatan ini dipilih karena relatif robust terhadap outlier dan sesuai untuk dataset dengan distribusi yang tidak selalu normal.

Kolom `CustomerID` dihapus karena tidak memiliki nilai prediktif terhadap churn.

Dataset hasil preprocessing siap digunakan untuk tahap feature encoding dan modeling.

## 12. Feature Encoding & Train-Test Split

## Imbalanced Data Handling

Dataset churn memiliki distribusi kelas yang tidak seimbang, di mana jumlah pelanggan non-churn lebih besar dibandingkan pelanggan churn. Kondisi ini berpotensi menyebabkan model bias terhadap kelas mayoritas.

Pada tahap ini dilakukan eksperimen beberapa teknik penanganan data tidak seimbang, yaitu:

- Undersampling (NearMiss)  
- Oversampling (SMOTE)  
- Hybrid sampling (SMOTE-ENN)  
- Class Weight  

Setiap teknik akan diterapkan secara terpisah dan dievaluasi menggunakan F1-score melalui cross-validation untuk menentukan metode yang memberikan performa terbaik.


```python
from imblearn.under_sampling import NearMiss
from imblearn.over_sampling import SMOTE
from imblearn.combine import SMOTEENN
from imblearn.pipeline import Pipeline as ImbPipeline
from sklearn.model_selection import cross_val_score
from xgboost import XGBClassifier

imbalance_results = []

sampling_methods = {
    "No Sampling": None,
    "NearMiss": NearMiss(),
    "SMOTE": SMOTE(random_state=42),
    "SMOTE-ENN": SMOTEENN(random_state=42)
}

base_model = XGBClassifier(
    random_state=42,
    eval_metric="logloss",
    use_label_encoder=False
)

for name, sampler in sampling_methods.items():
    
    steps = [
        ("preprocessor", preprocessor)
    ]
    
    if sampler:
        steps.append(("sampler", sampler))
    
    steps.append(("model", base_model))
    
    pipeline = ImbPipeline(steps=steps)
    
    scores = cross_val_score(
        pipeline,
        X_train,
        y_train,
        cv=5,
        scoring="f1"
    )
    
    imbalance_results.append({
        "Sampling Method": name,
        "Mean F1-score": scores.mean(),
        "Std F1-score": scores.std()
    })

imbalance_df = pd.DataFrame(imbalance_results).sort_values(
    by="Mean F1-score", ascending=False
)

imbalance_df
```

### Imbalanced Handling Result

Berdasarkan hasil eksperimen, metode **No Sampling** menghasilkan mean F1-score tertinggi (0.87), diikuti oleh SMOTE (0.86), SMOTE-ENN (0.75), dan NearMiss (0.42).

Hasil ini menunjukkan bahwa model XGBoost secara alami sudah mampu menangani ketidakseimbangan kelas dengan baik tanpa memerlukan teknik resampling tambahan. Penerapan oversampling maupun undersampling justru menurunkan performa model, yang mengindikasikan potensi munculnya noise atau distorsi distribusi data.

### Decision

Berdasarkan hasil tersebut, **metode No Sampling dipilih sebagai pendekatan final** untuk tahap modeling selanjutnya.

Keputusan ini diambil untuk menjaga data tetap merepresentasikan distribusi asli, serta menghindari risiko degradasi performa akibat data sintetis atau penghapusan data mayoritas.

## 13. Model Benchmarking

Pada tahap ini dilakukan pengujian beberapa algoritma klasifikasi sebagai kandidat model untuk memprediksi churn. Tujuan benchmarking adalah membandingkan performa berbagai model menggunakan **F1-score** sebagai metrik utama, karena F1-score mempertimbangkan precision dan recall secara bersamaan sehingga mampu menangkap trade-off antara False Positive dan False Negative.

Untuk memastikan perbandingan yang adil, digunakan dua jenis pipeline preprocessing:
- Pipeline dengan scaling untuk model berbasis jarak dan linear (Logistic Regression, KNN)
- Pipeline tanpa scaling untuk model berbasis tree (Decision Tree, Random Forest, Boosting)


```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import f1_score
from sklearn.compose import ColumnTransformer

# Preprocessor WITH scaling (for LogReg & KNN)
preprocessor_scaled = ColumnTransformer(
    transformers=[
        ("cat", OneHotEncoder(handle_unknown="ignore"), cat_cols),
        ("num", StandardScaler(), num_cols)
    ]
)

# Preprocessor WITHOUT scaling (for tree-based models)
preprocessor_noscale = preprocessor

models = {
    "Logistic Regression": (
        LogisticRegression(random_state=42),
        preprocessor_scaled
    ),
    "KNN": (
        KNeighborsClassifier(),
        preprocessor_scaled
    ),
    "Decision Tree": (
        DecisionTreeClassifier(random_state=42),
        preprocessor_noscale
    ),
    "Random Forest": (
        RandomForestClassifier(random_state=42),
        preprocessor_noscale
    ),
    "XGBoost": (
        XGBClassifier(
            random_state=42,
            eval_metric="logloss",
            use_label_encoder=False
        ),
        preprocessor_noscale
    )
}

results = []

for name, (model, prep) in models.items():
    pipeline = Pipeline(steps=[
        ("preprocessor", prep),
        ("model", model)
    ])
    
    pipeline.fit(X_train, y_train)
    y_pred = pipeline.predict(X_test)
    
    f1 = f1_score(y_test, y_pred)
    
    results.append({
        "Model": name,
        "F1-score": f1
    })

benchmark_df = pd.DataFrame(results).sort_values(
    by="F1-score", ascending=False
)

benchmark_df
```

### Benchmarking Result

Berdasarkan hasil benchmarking menggunakan F1-score, model **XGBoost** menghasilkan performa terbaik (0.98), diikuti oleh Random Forest (0.93), Decision Tree (0.92), KNN (0.74), dan Logistic Regression (0.61).

Hasil ini menunjukkan bahwa, pada baseline, model ensemble berbasis tree, khususnya boosting dan bagging, memiliki kemampuan yang lebih baik dalam menangkap pola kompleks dan non-linear pada data churn dibandingkan model linear dan distance-based.

## 14. Cross-Validation

Cross-validation dilakukan untuk mengevaluasi kestabilan performa seluruh model dan memastikan bahwa hasil evaluasi tidak hanya bergantung pada satu kali pembagian data (train-test split).  

Pada tahap ini digunakan 5-fold cross-validation dengan **F1-score** sebagai metrik utama, sesuai dengan tujuan untuk menyeimbangkan dampak False Positive dan False Negative.


```python
from sklearn.model_selection import cross_val_score
import numpy as np

cv_results = []

for name, (model, prep) in models.items():
    pipeline = Pipeline(steps=[
        ("preprocessor", prep),
        ("model", model)
    ])
    
    scores = cross_val_score(
        pipeline,
        X_train,
        y_train,
        cv=5,
        scoring="f1"
    )
    
    cv_results.append({
        "Model": name,
        "Mean F1-score": scores.mean(),
        "Std F1-score": scores.std()
    })

cv_df = pd.DataFrame(cv_results).sort_values(
    by="Mean F1-score", ascending=False
)

cv_df
```

### Cross-Validation Result

Berdasarkan hasil 5-fold cross-validation menggunakan F1-score, **XGBoost** menghasilkan performa terbaik (mean F1-score = 0.87), diikuti oleh Random Forest (0.84) dan Decision Tree (0.81). KNN dan Logistic Regression menunjukkan performa yang lebih rendah.

Selain itu, nilai standard deviation yang relatif kecil pada XGBoost dan Random Forest mengindikasikan bahwa kedua model tersebut memiliki performa yang cukup stabil di berbagai subset data.

Seluruh model selanjutnya akan dilanjutkan ke tahap hyperparameter tuning untuk memastikan perbandingan dilakukan secara adil sebelum pemilihan model final.

## 15. Hyperparameter Tuning

Pada tahap ini dilakukan hyperparameter tuning menggunakan GridSearchCV untuk seluruh model guna memperoleh konfigurasi parameter terbaik.  

Metrik utama yang digunakan adalah **F1-score**, sehingga proses tuning diarahkan untuk menyeimbangkan dampak False Positive dan False Negative.

Setiap model menggunakan pipeline preprocessing yang sama seperti pada tahap benchmarking, yaitu pipeline dengan scaling untuk Logistic Regression dan KNN, serta pipeline tanpa scaling untuk model berbasis tree dan boosting.


```python
from sklearn.model_selection import GridSearchCV

tuning_results = []

param_grids = {
    "Logistic Regression": {
        "model__C": [0.1, 1, 10]
    },
    "KNN": {
        "model__n_neighbors": [3,5,7,9]
    },
    "Decision Tree": {
        "model__max_depth": [3,5,7,10,None]
    },
    "Random Forest": {
        "model__n_estimators": [100,200],
        "model__max_depth": [5,10,None]
    },
    "XGBoost": {
        "model__n_estimators": [100,200],
        "model__max_depth": [3,5,7]
    }
}

for name, (model, prep) in models.items():
    
    pipeline = Pipeline(steps=[
        ("preprocessor", prep),
        ("model", model)
    ])
    
    grid = GridSearchCV(
        pipeline,
        param_grids[name],
        cv=5,
        scoring="f1",
        n_jobs=-1
    )
    
    grid.fit(X_train, y_train)
    
    tuning_results.append({
        "Model": name,
        "Best Mean F1-score": grid.best_score_,
        "Best Params": grid.best_params_
    })

tuning_df = pd.DataFrame(tuning_results).sort_values(
    by="Best Mean F1-score", ascending=False
)

tuning_df
```

### Hyperparameter Tuning Result

Hasil hyperparameter tuning menunjukkan bahwa **XGBoost** menghasilkan nilai mean F1-score tertinggi setelah optimasi parameter, diikuti oleh Random Forest dan Decision Tree.  

Urutan performa model setelah tuning konsisten dengan hasil benchmarking dan cross-validation, yang mengindikasikan bahwa pipeline eksperimen stabil dan reliabel.

Berdasarkan hasil ini, model dengan mean F1-score tertinggi setelah tuning akan dipilih sebagai model final.

## 16. Final Model Selection

Pada tahap ini dilakukan pemilihan model final berdasarkan hasil hyperparameter tuning menggunakan metrik utama F1-score. Model dengan nilai mean F1-score tertinggi dipilih sebagai model final karena memberikan keseimbangan terbaik antara precision dan recall.


```python
# Ambil model terbaik dari hasil tuning (XGBoost)
best_model_row = tuning_df.iloc[0]
best_model_row
```

```python
# Refit XGBoost dengan parameter terbaik
best_xgb_params = best_model_row["Best Params"]

final_model = Pipeline(steps=[
    ("preprocessor", preprocessor_noscale),
    ("model", XGBClassifier(
        random_state=42,
        eval_metric="logloss",
        use_label_encoder=False,
        **{k.replace("model__", ""): v for k, v in best_xgb_params.items()}
    ))
])

final_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay

# Predict on test set
y_pred_final = final_model.predict(X_test)

# Classification report
print(classification_report(y_test, y_pred_final))

# Confusion matrix
cm = confusion_matrix(y_test, y_pred_final)
ConfusionMatrixDisplay(cm).plot()
plt.show()
```

- Untuk kelas **churn (1)**, model mencapai **precision 0.99** dan **recall 0.96** dengan **F1-score sebesar 0.98**.

Confusion matrix menunjukkan bahwa dari 190 pelanggan churn aktual, sebanyak 182 pelanggan berhasil terdeteksi dengan benar (True Positive), sementara hanya 8 pelanggan churn yang tidak terdeteksi (False Negative).  

Selain itu, hanya terdapat 1 False Positive, yaitu pelanggan yang diprediksi churn padahal sebenarnya tidak.

Hasil ini mengindikasikan bahwa model mampu meminimalkan baik False Negative maupun False Positive, sehingga sejalan dengan tujuan bisnis untuk menekan kehilangan pelanggan sekaligus menghindari biaya retensi yang tidak perlu.

## 18. Feature Importance & Model Interpretability

Pada tahap ini dilakukan interpretasi model final menggunakan SHAP (SHapley Additive exPlanations) untuk memahami kontribusi masing-masing fitur terhadap prediksi churn.

SHAP dipilih karena mampu memberikan interpretasi yang konsisten dan berbasis teori game, serta dapat menjelaskan pengaruh fitur baik secara global maupun lokal. Dengan demikian, hasil interpretasi ini membantu memastikan bahwa model tidak bersifat black-box dan tetap dapat dipahami secara bisnis.


```python
import shap

# Transform training data using fitted preprocessor
X_train_transformed = final_model.named_steps["preprocessor"].transform(X_train)

# Get feature names after preprocessing
feature_names = final_model.named_steps["preprocessor"].get_feature_names_out()

# Extract trained model
xgb_model = final_model.named_steps["model"]

# Create SHAP explainer
explainer = shap.TreeExplainer(xgb_model)

# Compute SHAP values
shap_values = explainer.shap_values(X_train_transformed)

# Global feature importance (bar plot)
shap.summary_plot(
    shap_values,
    X_train_transformed,
    feature_names=feature_names,
    plot_type="bar"
)

# Detailed SHAP summary plot
shap.summary_plot(
    shap_values,
    X_train_transformed,
    feature_names=feature_names
)
```

### Implikasi utama:

`num__Tenure` menjadi faktor paling dominan, menunjukkan bahwa lamanya hubungan pelanggan dengan perusahaan adalah penentu utama risiko churn.

Faktor pengalaman layanan seperti `num__Complain` dan `num__SatisfactionScore` juga memiliki kontribusi besar.
